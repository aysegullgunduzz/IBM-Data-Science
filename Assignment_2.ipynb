```python
import yfinance as yf
import matplotlib.pyplot as plt

# Function to create a graph (assuming make_graph is predefined)
def make_graph(stock_data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(stock_data['Date'], stock_data['Close'])
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.show()

# Download Tesla stock data
tesla_data = yf.download('TSLA', start='2020-01-01', end='2023-01-01')

# Reset the index to make 'Date' a column
tesla_data.reset_index(inplace=True)

# Call the make_graph function with the Tesla stock data
make_graph(tesla_data, 'Tesla Stock Prices')
```
